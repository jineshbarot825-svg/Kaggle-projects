### 1. ðŸš€ The Pitch (Enterprise Agents Track)

**The Problem:** In any software development team, code review is a critical but time-consuming bottleneck. Senior developers spend hours manually checking new code for bugs, style violations, and security flaws. This process is slow, expensive, and prone to human inconsistency.

**The Solution:** "The 'Dev-Guardian' Swarm" is an autonomous **enterprise agent** team that automates this entire workflow. It runs as a long-running "daemon" that monitors a (mocked) code submission queue.

When new code is "submitted," the swarm activates:
1.  A **Sequential** `PlannerAgent` logs the submission.
2.  A **Parallel Team** of specialist agents (`SecurityAgent`, `StyleAgent`, `TestAgent`) review the code simultaneously.
3.  The `StyleAgent` **Reads from Long-Term Memory** to check the code against "Company Coding Standards."
4.  The `TestAgent` uses a (mocked) **Code Execution Tool** to run unit tests.
5.  An **Evaluation Agent** (the `LeadDeveloperAgent`) acts as the "brain." It reads all the technical reports from the parallel team and *generates* a single, polite, human-readable summary.
6.  A final agent uses a **Real Custom Tool** to write this synthesized review to a formal `CODE_REVIEW_REPORT.md` file.

This project automates a complex, real-world business workflow, freeing up senior developers to focus on architecture and new features.

### 2. âš™ï¸ Features Demonstrated

This notebook is a capstone project that demonstrates all key concepts from the 5-Day AI Agents Intensive Course:

1.  **Multi-agent System:**
    * **Sequential Agents:** `PlannerAgent` -> (Parallel Team) -> `LeadDeveloperAgent` -> `PublisherAgent`.
    * **Parallel Agents:** `SecurityAgent`, `StyleAgent`, and `TestAgent` all run at the same time.
    * **Loop Agent:** The main `run_dev_guardian_swarm` function acts as a "daemon" loop, processing a queue of submissions.

2.  **Tools:**
    * **Mocked Built-in Tool (Google Search):** `tool_google_search` is used by the `SecurityAgent` to find vulnerabilities.
    * **Mocked Built-in Tool (Code Execution):** `tool_run_unit_tests` is used by the `TestAgent`.
    * **Real Custom Tool:** `tool_save_review_to_markdown` saves the final, formatted review to a `.md` file.

3.  **Long-Running Operations (Pause/Resume):**
    * The main swarm function is a **Loop** that "wakes up," processes the submission queue, and "pauses" (`time.sleep`) to wait for new code.

4.  **Long-Term Memory:**
    * **Write:** The `HistorianAgent` saves the final, synthesized review to the `MEMORY_BANK` for auditing.
    * **Read:** The `MEMORY_BANK` is pre-loaded with "Company Coding Standards," which the `StyleAgent` **reads** to perform its review.

5.  **Agent Evaluation (The "Brain"):**
    * The `LeadDeveloperAgent` is the "brain." It reads and evaluates three different technical reports, *synthesizes* them, and *generates* a new, human-readable summary.

6.  **Observability:**
    * A `log_message()` function provides a rich, real-time log of every agent's actions and decisions.

### 3. ðŸ¤– System Architecture

This swarm operates on a sophisticated Sequential -> Parallel -> Sequential workflow, which ensures both efficiency and proper evaluation.

Here is the step-by-step data and logic flow:

1.  **[LOOP START]** (`run_dev_guardian_swarm`)
    * `WatcherAgent` monitors the (mocked) code submission queue.

2.  **[SEQUENTIAL 1]** (`PlannerAgent`)
    * Receives one code submission from the queue.
    * Logs the task and prepares the data.

3.  **[PARALLEL TEAM]** (Specialists)
    * The `PlannerAgent` passes the code to three specialists who run at the *same time*:
        * `SecurityAgent`: Checks for vulnerabilities (uses `tool_google_search`).
        * `StyleAgent`: Checks for style errors (uses `MEMORY_BANK` to read "Coding Standards").
        * `TestAgent`: Runs unit tests (uses `tool_run_unit_tests`).

4.  **[SEQUENTIAL 2]** (`LeadDeveloperAgent` - The "Brain")
    * Waits for all three parallel agents to finish.
    * **Evaluates** and **synthesizes** their three separate reports into one final, human-readable review.

5.  **[SEQUENTIAL 3]** (`HistorianAgent` & `PublisherAgent`)
    * `HistorianAgent`: **Writes** the final review to the `MEMORY_BANK` for auditing.
    * `PublisherAgent`: Uses the **Custom Tool** (`tool_save_review_to_markdown`) to save the final report to `CODE_REVIEW_REPORT.md`.

6.  **[LOOP END]**
    * The swarm **pauses** (`time.sleep`) and waits for the next submission.

# **ðŸ“š Imports & Libraries**

```python
import time
import json
import datetime
import random
import os
import textwrap # Used for formatting our final Markdown report

print("Cell 1: Libraries imported successfully.")
```

# ðŸ§  Global Setup (Memory Bank & Logger)

```python
MEMORY_BANK = {
    "coding_standards": {},
    "review_history": [],
    "alerts_fired": 0,
}

# This helper function will act as our Observability/Logging
def log_message(agent_name, message, level="INFO"):
    """Helper function for formatted logging."""
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{timestamp}] [{agent_name}] [{level}]: {message}")

print("Cell 2: Global Memory Bank and Logger initialized.")
```

# ðŸ’¾Pre-load Memory (Company Coding Standards)
This is a critical step. We are pre-loading the Long-Term Memory with rules for our agents to read.

```python
MEMORY_BANK["coding_standards"] = {
    "max_line_length": 88,
    "disallowed_functions": ["exec", "eval"],
    "required_docstrings": True
}

print("Cell 3: Long-Term Memory pre-loaded with 'Company Coding Standards'.")
print(json.dumps(MEMORY_BANK["coding_standards"], indent=2))
```

# ðŸ› ï¸Agent Toolbox (Mocked & Custom Tools)

```python
# --- Mocked Built-in Tool (Google Search) ---
def tool_google_search(query: str) -> str:
    """MOCK BUILT-IN TOOL: Simulates a Google Search for vulnerabilities."""
    log_message("GoogleSearchTool", f"Searching for: '{query}'")
    vulnerabilities = {
        "exec": "High-risk: 'exec' can lead to arbitrary code execution.",
        "eval": "High-risk: 'eval' is dangerous and can execute any string as code.",
        "os.system": "Medium-risk: 'os.system' can be used for shell injection."
    }
    for func in vulnerabilities:
        if func in query:
            return json.dumps([{"source": "security-blog.com", "finding": vulnerabilities[func]}])
    return json.dumps([{"source": "security-blog.com", "finding": "No critical vulnerabilities found."}])

# --- Mocked Built-in Tool (Code Execution) ---
def tool_run_unit_tests(code_snippet: str) -> str:
    """MOCK BUILT-IN TOOL (CODE EXEC): Simulates running unit tests."""
    log_message("TestRunnerTool", "Simulating unit tests...")
    
    # Mock logic: if the code is simple, it passes.
    status = "PASS"
    if "def" not in code_snippet or "return" not in code_snippet:
        status = "FAIL"
        
    return json.dumps({
        "status": status,
        "tests_run": 5,
        "tests_passed": 4 if status == "PASS" else 2,
        "tests_failed": 1 if status == "PASS" else 3
    })

# --- REAL Custom Tool (Markdown File Saver) ---
def tool_save_review_to_markdown(report_content: str, filename: str):
    """
    REAL CUSTOM TOOL: Saves the final, formatted review to a Markdown file.
    """
    log_message("CustomFileTool", f"Saving final review to '{filename}'")
    try:
        # textwrap.dedent removes leading whitespace, making our report clean
        with open(filename, 'a') as f:
            f.write(textwrap.dedent(report_content))
            f.write("\n\n" + "="*80 + "\n\n")
        log_message("CustomFileTool", "Review successfully saved.")
        MEMORY_BANK["alerts_fired"] += 1
    except Exception as e:
        log_message("CustomFileTool", f"Failed to save file: {e}", level="ERROR")

print("Cell 4: All tools (mocked and custom) are defined.")
```

# ðŸ¤– The Agent Team: Part 1 (The Specialists)
These are the parallel agents who perform the first-pass review.

```python
# --- Sequential Agent 1: The Planner ---
def planner_agent(submission_id: int, code_snippet: str) -> dict:
    """SEQUENTIAL AGENT 1: Logs the new submission."""
    log_message("PlannerAgent", f"New code submission {submission_id} received.")
    return {"submission_id": submission_id, "code_snippet": code_snippet}

# --- Parallel Agent 1: The Security Agent ---
def security_agent(code_snippet: str) -> dict:
    """PARALLEL AGENT 1: Checks for security vulnerabilities."""
    log_message("SecurityAgent", "Analyzing code for security flaws...")
    # READS from Memory to get the list of disallowed functions
    disallowed = MEMORY_BANK["coding_standards"]["disallowed_functions"]
    findings = []
    
    for func in disallowed:
        if func in code_snippet:
            log_message("SecurityAgent", f"Found high-risk function: '{func}'", level="WARN")
            # Use the tool to get more info
            details = json.loads(tool_google_search(f"vulnerability for '{func}'"))
            findings.append(f"High-Risk Function: `{func}`. Finding: {details[0]['finding']}")
            
    log_message("SecurityAgent", f"Security review complete. {len(findings)} issues found.")
    return {"security_report": findings}

# --- Parallel Agent 2: The Style Agent ---
def style_agent(code_snippet: str) -> dict:
    """PARALLEL AGENT 2: Checks for style violations by READING from memory."""
    log_message("StyleAgent", "Analyzing code for style violations...")
    
    # READS from Long-Term Memory
    standards = MEMORY_BANK["coding_standards"]
    findings = []
    
    # Mock logic
    if standards["required_docstrings"] and '"""' not in code_snippet:
        findings.append("Style Error: Missing required docstring.")
    
    for i, line in enumerate(code_snippet.split('\n')):
        if len(line) > standards["max_line_length"]:
            findings.append(f"Style Error: Line {i+1} exceeds {standards['max_line_length']} characters.")
            
    log_message("StyleAgent", f"Style review complete. {len(findings)} issues found.")
    return {"style_report": findings}

# --- Parallel Agent 3: The Test Agent ---
def test_agent(code_snippet: str) -> dict:
    """PARALLEL AGENT 3: "Runs" unit tests using the Code Execution tool."""
    log_message("TestAgent", "Preparing to run unit tests...")
    
    # Uses the (mocked) Code Execution tool
    test_results = json.loads(tool_run_unit_tests(code_snippet))
    
    log_message("TestAgent", f"Unit tests complete. Status: {test_results['status']}")
    return {"test_report": test_results}

print("Cell 5: Planner and Parallel Agents defined.")
```

# ðŸ§  The Agent Team: Part 2 (The "Brain")
This is the Evaluation Agent that synthesizes the reports and the Publisher Agent that saves the file.

```python
# --- Evaluation Agent: The "Brain" ---
def lead_developer_agent(reports: dict, submission_id: int) -> dict:
    """
    EVALUATION AGENT: The "brain."
    Synthesizes reports from the parallel team into one human-readable review.
    Demonstrates: Agent Evaluation, Generative AI (Mocked)
    """
    log_message("LeadDeveloperAgent", f"Synthesizing all reviews for submission {submission_id}...")
    
    security_report = reports["security_report"]
    style_report = reports["style_report"]
    test_report = reports["test_report"]
    
    # --- MOCKED GENERATIVE AI ---
    # This is where an LLM would write a polite, human-readable summary.
    
    final_review = f"""
    ### **Code Review for Submission #{submission_id}**
    
    **Overall Status:** {'NEEDS REVISION' if security_report or style_report or test_report['status'] == 'FAIL' else 'APPROVED'}
    
    Hi team,
    
    Here is the automated review for your latest submission.
    
    ---
    
    #### **ðŸ§ª Test Results**
    * **Status:** {test_report['status']}
    * **Summary:** {test_report['tests_passed']} / {test_report['tests_run']} tests passed.
    
    """
    
    if security_report:
        final_review += "#### **ðŸ”’ Security Review (High Priority)**\n"
        for finding in security_report:
            final_review += f"* {finding}\n"
    
    if style_report:
        final_review += "\n#### **ðŸŽ¨ Style Review**\n"
        for finding in style_report:
            final_review += f"* {finding}\n"
            
    if not security_report and not style_report and test_report['status'] == 'PASS':
        final_review += "\n**Conclusion:** Great work! All checks passed. This is ready to merge.\n"
    else:
        final_review += "\n**Conclusion:** Please address the high-priority security and style issues above.\n"
        
    log_message("LeadDeveloperAgent", "Synthesis complete. Handing off to Publisher.")
    return {"final_review": final_review, "status": "Ready for Publishing"}

# --- Sequential Agent 3: The Historian ---
def historian_agent(submission_id: int, final_review: str):
    """SEQUENTIAL AGENT 3: Saves the final review to Long-Term Memory (Write)."""
    log_message("HistorianAgent", f"Saving review for {submission_id} to Memory Bank.")
    
    # WRITE to memory
    MEMORY_BANK["review_history"].append({
        "id": submission_id,
        "timestamp": datetime.datetime.now().isoformat(),
        "review": final_review
    })
    log_message("HistorianAgent", "Review saved to memory.")

# --- Sequential Agent 4: The Publisher ---
def publisher_agent(final_review: str):
    """SEQUENTIAL AGENT 4: Uses the Custom Tool to save the final report."""
    log_message("PublisherAgent", "Calling Custom Tool to save Markdown file...")
    
    # Use the REAL Custom Tool
    tool_save_review_to_markdown(final_review, "CODE_REVIEW_REPORT.md")

print("Cell 6: Evaluation ('Brain') and Publisher Agents defined.")
```

# ðŸ”„ The "Daemon" Loop (Main Function)
This is the main "Loop Agent" that simulates a server, processing a queue of code submissions.

```python
def run_dev_guardian_swarm(loop_delay_seconds: int):
    """
    Main function to run the agent swarm (Loop Agent).
    Simulates a "daemon" processing a queue of submissions.
    Demonstrates: Loop, Long-Running Ops (Pause/Resume), Sequential, Parallel
    """
    
    log_message("DevGuardianSwarm", "Swarm is active. Monitoring code submission queue...")
    
    # --- MOCKED CODE SUBMISSION QUEUE ---
    # This simulates a queue of code waiting for review.
    code_submission_queue = [
        {
            "id": 101,
            # --- FIX: Using triple quotes for readability ---
            "code": """def add(a, b):
    # This line is intentionally too long to fail the style check
    print('This function is going to add two numbers together and it is a very long line')
    return a + b"""
        },
        {
            "id": 102,
            # --- FIX: Using triple quotes to fix the SyntaxError ---
            "code": """import os
def run_command(cmd):
    '''This is a docstring'''
    # This function uses a dangerous function!
    exec(cmd)
    return True"""
        },
        {
            "id": 103,
            # --- FIX: Using triple quotes for consistency ---
            "code": '''"""
This function is perfect.
"""
def subtract(a, b):
    return a - b'''
        }
    ]
    
    # --- Clear alert file at the start ---
    alert_filename = "CODE_REVIEW_REPORT.md"
    if os.path.exists(alert_filename):
        os.remove(alert_filename)
        log_message("DevGuardianSwarm", "Cleared old review file.")

    # --- Phase 1: The "Daemon" (Loop) ---
    loop_count = 0
    while code_submission_queue:
        loop_count += 1
        log_message("WatcherAgent", f"--- Starting Loop {loop_count} ---")
        
        # Get the next item from the queue
        submission = code_submission_queue.pop(0)
        
        # --- Phase 2: Planning (Sequential) ---
        planned_task = planner_agent(submission["id"], submission["code"])
        
        # --- Phase 3: Parallel Execution ---
        log_message("WatcherAgent", f"Spawning parallel agents for submission {submission['id']}...")
        
        report_part_1 = security_agent(planned_task["code_snippet"])
        report_part_2 = style_agent(planned_task["code_snippet"])
        report_part_3 = test_agent(planned_task["code_snippet"])
        
        # Combine parallel results
        combined_reports = {**report_part_1, **report_part_2, **report_part_3}
        
        # --- Phase 4: Evaluation (Sequential "Brain") ---
        evaluation = lead_developer_agent(combined_reports, planned_task["submission_id"])
        
        # --- Phase 5: Memory (Write) & Publishing (Custom Tool) ---
        historian_agent(planned_task["submission_id"], evaluation["final_review"])
        publisher_agent(evaluation["final_review"])
        
        # --- Phase 6: LONG-RUNNING OPERATION (PAUSE/RESUME) ---
        if code_submission_queue:
            log_message("WatcherAgent", f"Operation complete. Pausing for {loop_delay_seconds}s...")
            time.sleep(loop_delay_seconds)
            log_message("WatcherAgent", "Resuming operation. Checking for new code...")
    
    log_message("DevGuardianSwarm", "All submissions processed. Swarm is shutting down.")

print("Cell 7: Main 'Daemon' loop function defined.")
```

# ðŸš€ Run the Swarm!
This cell will execute the entire swarm. You will see the logs print out in real-time as it processes all three (mocked) code submissions.

```python
# --- Configuration ---\
LOOP_DELAY_SECONDS = 3 # How long to "pause" (3s for demo)

# --- Clear the memory from previous runs (but keep the standards!) ---
MEMORY_BANK["review_history"] = []
MEMORY_BANK["alerts_fired"] = 0
print("--- Cleared Memory Bank history for new run ---")

# --- Run the entire agent swarm ---\
run_dev_guardian_swarm(
    loop_delay_seconds=LOOP_DELAY_SECONDS
)

print("\n\n" + "="*40)
print("âœ… SWARM RUN COMPLETE âœ…")
print(f"Total Reviews Generated: {MEMORY_BANK['alerts_fired']}")
print("="*40)
```
