## 1. The Pitch 

### 1.1. The Problem (Core Concept & Value)

The internet has endless information, but no roadmap. When a person wants to learn a complex topic (like "Quantum Computing" or "Rust Programming"), they face a "cold start" problem:
* What do I learn first?
* What are the best resources for *my* background?
* How do I know I'm actually learning?

This aimless searching is frustrating, inefficient, and causes many to give up.

### 1.2. The Solution: `Learn-Navigator` (Core Concept & Value)

**`Learn-Navigator` is a multi-agent system that acts as an expert curriculum architect.**

The user provides three simple inputs:
1.  **Topic:** (e.g., "Rust Programming")
2.  **Background:** (e.g., "I'm a Python developer")
3.  **Learning Style:** (e.g., "I'm a visual learner, I prefer videos")

The agent team then researches, designs, and validates a complete, personalized, week-by-week learning plan, complete with resources and quizzes.

### 1.3. The Value Proposition

This agent system saves a self-learner dozens of hours of research and provides a validated, high-quality path to mastery. It democratizes education by providing a "personal tutor" experience, tailored to each user's unique needs.

## 2. The Implementation 

### 2.1. System Architecture

The system is managed by a `Chief-Tutor-Agent` which coordinates a team of three specialist agents to fulfill the user's request.

1.  The `Chief-Tutor-Agent` receives the user's request and stores it in **Session Memory**.
2.  It dispatches the `Research-Agent` **(Parallel task)**.
3.  The `Research-Agent` uses **Google Search** (built-in tool), **YouTube Search** (OpenAPI tool), and a **Web Scraper** (custom tool) to find resources.
4.  The `Research-Agent` performs **Context Engineering** by summarizing the resources.
5.  The `Chief-Tutor-Agent` passes the research to the `Curriculum-Agent` **(Sequential task)**.
6.  The `Curriculum-Agent` designs the curriculum. For each module, it uses **A2A Protocol** to *ask* the `Quiz-Agent` to generate questions.
7.  The `Quiz-Agent` uses the **Code Execution** (built-in tool) to create coding challenges.
8.  The final plan is presented. The user's feedback can trigger a **Loop** to refine the search.
9.  The final plan is saved to **Long-term Memory (Memory Bank)**.

# Cell 5: Step 2.2 - Defining the Agents

```python
# Cell 5: Step 2.2 - Defining the Agents

log.info("Defining Agents...")

# --- Agent 1: Research-Agent ---
research_agent = Agent(
    name="Research-Agent",
    description="Finds and summarizes resources.",
    prompt="""
    You are a world-class researcher. Your goal is to find the BEST 
    articles and videos for a user's learning query.
    You MUST use your tools.
    
    1.  Use `Google Search` to find 3-5 high-quality articles.
    2.  For each article, use `scrape_article_text` to get its content.
    3.  If the user is a 'visual learner', use `Youtube` to find 3 relevant videos.
    4.  Compile all findings (summaries and video links) into a JSON object.
    """,
    # We pass in the *fake tool objects* we defined
    tools=[google_search_tool, youtube_search_tool, scrape_article_text_tool],
    model="gemini-1.5-pro" # Hits "Use of Gemini" bonus
)
log.info(f"Defined Agent: {research_agent.name}")

# --- Agent 2: Quiz-Agent ---
quiz_agent = Agent(
    name="Quiz-Agent",
    description="Generates quizzes for a topic.",
    prompt="""
    You are an expert test creator. Given a specific topic (e.g., 'Rust Ownership'), 
    create 3 high-quality questions to validate understanding.
    
    1.  Create 2 multiple-choice questions.
    2.  Create 1 coding challenge. Use your `code_execution` tool to 
        provide a skeleton and a test case.
    3.  Return the questions in a JSON format.
    """,
    tools=[code_execution_tool],
    model="gemini-1.5-pro"
)
log.info(f"Defined Agent: {quiz_agent.name}")

# --- Agent 3: Curriculum-Agent ---
curriculum_agent = Agent(
    name="Curriculum-Agent",
    description="Designs the final learning plan.",
    prompt="""
    You are an expert curriculum designer. You will be given a user's goal,
    their background, and a JSON object of research.
    
    Your job is to create a week-by-week learning plan (e.g., 4 weeks).
    For each week, define:
    1.  The `topic` (e.g., "Week 1: Rust Basics & Syntax").
    2.  `learning_objectives` (3-5 bullet points).
    3.  `resources` (link the articles and videos from the research).
    
    IMPORTANT: After you design each week, you MUST call the `Quiz-Agent`
    using A2A (Agent-to-Agent) protocol to generate a quiz for that week's topic.
    
    Format the entire plan as a single, detailed Markdown string.
    """,
    tools=[], # This agent orchestrates, it doesn't *use* tools itself, it calls other agents
    model="gemini-1.5-pro"
)
log.info(f"Defined Agent: {curriculum_agent.name}")
```

# Cell 6: Step 2.3 - Building the System (The Chief-Tutor)

```python
# Cell 6: Step 2.3 - Building the System (The Chief-Tutor)

log.info("Defining the Main System (LearnNavigator class)...")

# This class represents our "Chief-Tutor" agent and main application.
# It demonstrates:
# - Multi-agent system (Sequential & Parallel)
# - Sessions & Memory (State)
# - Long-term Memory (Memory Bank)
# - A2A Protocol
# - Long-running operations (the `generate_plan` method)
# - Loop agents (based on feedback)

class LearnNavigator:
    def __init__(self):
        # We need a session service to manage state between agents
        self.session_service = InMemorySessionService()
        
        # This is for Long-term Memory
        self.memory_bank = MemoryBank() 
        log.info("LearnNavigator system initialized with FAKE SessionService and FAKE MemoryBank.")

    def generate_plan(self, topic: str, background: str, style: str, simulated_feedback: str = None):
        """
        (FAKE) This is the main "Long-Running Operation".
        It orchestrates the entire agent team.
        """
        log.info(f"--- NEW PLAN REQUESTED ---")
        log.info(f"Topic: {topic}, Background: {background}, Style: {style}")

        # 1. Create a new session for this request
        # This demonstrates Sessions & Memory (State)
        session_id = f"session_{hash(topic + background + style)}"
        session = self.session_service.create_session(session_id)
        session.set("topic", topic)
        session.set("background", background)
        session.set("style", style)
        
        # 2. Check Long-term Memory
        past_plan = self.memory_bank.get(f"plan_for_{topic}")
        if past_plan:
            log.info("Found a similar plan in Long-term Memory. Using as reference.")
            session.set("past_plan_reference", past_plan)

        # --- A2A PROTOCOL: Step 1 (Tutor -> Research) ---
        log.info(f"[A2A] Chief-Tutor dispatching Research-Agent...")
        research_query = f"Find learning resources for {topic} for a {background}. User is a {style} learner."
        
        # This is a PARALLEL task (in a real system, this would be non-blocking)
        # We call our FAKE run_agent function
        research_result = run_agent(research_agent, research_query, session=session)
        
        log.info(f"[A2A] Research-Agent returned FAKE findings: {research_result}")
        
        # --- A2A PROTOCOL: Step 2 (Tutor -> Curriculum) ---
        log.info(f"[A2A] Chief-Tutor dispatching Curriculum-Agent...")
        # The curriculum agent will internally call the Quiz-Agent (A2A)
        curriculum_query = f"""
        Create a 4-week curriculum for {topic} for a {background}.
        Use this research: {research_result}.
        
        This agent will internally call the Quiz-Agent.
        """
        
        final_plan = run_agent(curriculum_agent, curriculum_query, session=session)
        
        log.info(f"[A2A] Curriculum-Agent returned the FAKE final plan.")
        
        # --- DEMONSTRATION OF A LOOP ---
        if simulated_feedback:
            log.info(f"[Loop] Simulated user feedback received: '{simulated_feedback}'")
            if "too hard" in simulated_feedback:
                log.info("[Loop] User feedback requires a loop. Re-dispatching Research-Agent for 'easier' content...")
                # This is the loop!
                research_query_2 = f"Find *beginner* and *simpler* resources for {topic}, user found first pass too hard."
                easy_research = run_agent(research_agent, research_query_2, session=session)
                log.info(f"[Loop] Found new easy resources: {easy_research}")
                final_plan += "\n\n**[Revised Resources (from Loop)]**: Based on your feedback, here are some simpler resources: " + str(easy_research)

        
        # 3. Save to Long-term Memory
        self.memory_bank.save(f"plan_for_{topic}", final_plan)
        log.info(f"Plan saved to Long-term Memory.")
        
        return final_plan

# Initialize our system
learn_navigator = LearnNavigator()
```
